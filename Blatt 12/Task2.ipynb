```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## inititalize parameter

```python
G = 10
a = 2
m = 1
M = 2
```

```python
mu = (m*M)/(m+M)
mu
```

```python
k = np.arange(-np.pi/(2*a), np.pi/(2*a), 0.001)
k
```

$\omega^2 = \dfrac{G}{\mu} \pm G \sqrt{\left( \dfrac{1}{\mu} \right)^2 - \dfrac{4\sin^2{\left(ka\right)}}{mM}}$

## define dispersion relation

```python
def omega_plus(k):
    expr = G/mu + G*np.sqrt((1/mu)**2 - (4*np.sin(k*a)**2)/(m*M))
    return np.sqrt(expr)
```

```python
def omega_minus(k):
    expr = G/mu - G*np.sqrt((1/mu)**2 - (4*np.sin(k*a)**2)/(m*M))
    return np.sqrt(expr)
```

```python
sns.set_context("talk")
plt.figure(figsize=(12,6))
plt.plot(k, omega_plus(k), label=r"$\omega_+$ (optical branch)")
plt.plot(k, omega_minus(k), label=r"$\omega_-$ (acoustic branch)")
plt.xlabel("k")
plt.ylabel(r"$\omega$")
plt.legend()
#plt.savefig("dispersion.png", dpi=500)
plt.show()
```
