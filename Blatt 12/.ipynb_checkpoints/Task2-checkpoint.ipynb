```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

$\omega^2 \approx \dfrac{G}{\mu} \cdot \left[ 1 \pm \left(1 - 2\left( ka \right)^2\dfrac{\mu}{m+M}\right)\right]$

$\omega \approx \sqrt{\dfrac{G}{\mu} \cdot \left[ 1 \pm \left(1 - 2\left( ka \right)^2\dfrac{\mu}{m+M}\right)\right]}$

$ \omega_+ \approx \sqrt{\dfrac{2G}{\mu}} $ ------------------------------
$ \omega_+ = \sqrt{\dfrac{2G}{\mu}}  \sqrt{1-(ka)^2 \dfrac{\mu}{m+M}}$

$ \omega_- = \sqrt{\dfrac{2G\mu}{mM}} \cdot ka$

```python
G = 10
a = 2
m = 1
M = 2
```

```python
mu = (m*M)/(m+M)
mu
```

```python
k = np.arange(-np.pi/(2*a), np.pi/(2*a), 0.001)
k
```

```python
def omega_plus(k):
    expr = np.sqrt(2*G/mu)*np.sqrt(1 - (k*a)**2*(mu/(m+M)))
    return expr
```

```python
def omega_minus(k):
    expr = np.sqrt(((2*G*mu)/(m*M))*(k*a)**2)
    return expr
```

```python
sns.set_context("talk")
plt.figure(figsize=(12,6))
plt.plot(k, omega_plus(k), label=r"$\omega_+$ (optical branch)")
plt.plot(k, omega_minus(k), label=r"$\omega_-$ (acoustic branch)")
plt.xlabel("k")
plt.ylabel(r"$\omega$")
plt.legend()
plt.savefig("dispersion.png", dpi=500)
plt.show()
```
