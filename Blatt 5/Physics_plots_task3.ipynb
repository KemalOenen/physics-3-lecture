```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import h,m_e
```

$\sqrt{\dfrac{1}{2\pi \sigma^2(t)}}\exp{\left(\dfrac{-\left(x-\frac{\hbar k_0}{m}t \right)^2}{2\sigma^2(t)} \right)}$

$\sigma^2(t) = \dfrac{a^2}{4} \left(1 + \dfrac{4\hbar^2t^2}{m^2a^4} \right)$

```python
x = np.linspace(-10,10,100)
t1 = 0
t2 = -10**3
t3 = -10**4
t4 = -10**5
t5 = 10**3
t6 = 10**4
t7 = 10**5
a = 3
m = m_e
hbar = h/(2*np.pi)
k_0 = 1
```

```python
def sigma_squared(t):
    return ((a**2)/4) * (1 + ((4*(hbar**2)*(t**2))/((m**2) * (a**4))))
def psi(x,t):
    return np.sqrt(1/(2*np.pi*sigma_squared(t))) * np.exp(-((x-(hbar*k_0*(t/m)))**2)/(2*sigma_squared(t)))
```

```python
plt.plot(x, psi(x,t1), label = r't=0s')
plt.plot(x, psi(x,t3), label = r'$t=-10^{4}$s')
plt.plot(x, psi(x,t4), label = r'$t=-10^{5}$s')
plt.plot(x, psi(x,t6), label = r'$t=10^4$s')
plt.plot(x, psi(x,t7), label = r'$t=10^5$s')
plt.xlabel(r'x')
plt.ylabel(r'$\left|\psi(x,t) \right|^2$')
plt.legend()
plt.savefig('psi_xt.png', dpi=500)
plt.show()
```
